## RandomForest (RandomizedSearchCV)

```python
estimator = RandomForestClassifier()

para_depth = [depth for depth in range(10,40)]
para_split = [n_split *10 for n_split in range(1,4)]
para_leaf = [n_leaf * 2 for n_leaf in range(1,4)]

param_grid = {'max_depth': para_depth, 'min_samples_split': para_split, 'min_samples_leaf':para_leaf}

grid_rf = RandomizedSearchCV(estimator, param_grid, scoring = 'accuracy', n_jobs = -1)
grid_rf.fit(X_train, y_train)
```

```python
print("Best parameter model: \n{}".format(grid_rf.best_estimator_))
print("Best parameter: \n{}".format(grid_rf.best_params_))
print("Best score: \n{}".format(grid_rf.best_score_.round(4)))
```
